```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
plt.xkcd()
colors = ['#332288', '#CC6677', '#6699CC', '#117733']
import matplotlib 
matplotlib.rc('xtick', labelsize=15) 
matplotlib.rc('ytick', labelsize=15) 
matplotlib.rc('lines', linewidth=3)
```

```python
R, sqrt_vc2_tot, sqrt_vc2_dm, sqrt_vc2_disk, sqrt_vc2_bulge = np.loadtxt('../Dehnen_rotcurve.txt', unpack=True, skiprows=1)

fig, axis = plt.subplots(1, figsize=(5, 5))
axis.plot(R, sqrt_vc2_dm, color=colors[1], label='Dark Matter')
axis.plot(R, sqrt_vc2_disk, color=colors[2], label='Stars & Gas')
axis.plot(R[::40], sqrt_vc2_tot[::40], 'k.', label='Measured')
#axis.set_axis_bgcolor('red')
axis.set_xticks([])
axis.set_yticks([])
axis.set_xlim(0, 30)
axis.set_ylim(0, 110)
axis.legend(loc='lower right')
axis.set_xlabel('Distance from Galaxy Center', fontsize=15)
axis.set_ylabel('Circular Velocity', fontsize=15)

plt.tight_layout()
plt.savefig('./rotCurveAll.pdf', dpi=300)

fig, axis = plt.subplots(1, figsize=(5, 5))
axis.plot(R, sqrt_vc2_disk, color=colors[2], label='Stars & Gas')

axis.set_xticks([])
axis.set_yticks([])
axis.set_xlim(0, 30)
axis.set_ylim(0, 110)
axis.legend(loc='lower right')
axis.set_xlabel('Distance from Galaxy Center', fontsize=15)
axis.set_ylabel('Circular Velocity', fontsize=15)

plt.tight_layout()
plt.savefig('./rotCurveDisk.pdf', dpi=300)

fig, axis = plt.subplots(1, figsize=(5, 5))
axis.plot(R, sqrt_vc2_disk, color=colors[2], label='Stars & Gas')
axis.plot(R[::40], sqrt_vc2_tot[::40], 'k.', label='Measured')

axis.set_xticks([])
axis.set_yticks([])
axis.set_xlim(0, 30)
axis.set_ylim(0, 110)
axis.legend(loc='lower right')
axis.set_xlabel('Distance from Galaxy Center', fontsize=15)
axis.set_ylabel('Circular Velocity', fontsize=15)

plt.tight_layout()
plt.savefig('./rotCurveDiskMeasured.pdf', dpi=300)
```
