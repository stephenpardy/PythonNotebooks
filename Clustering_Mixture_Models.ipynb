```python
import numpy as np
from sklearn.cluster import DBSCAN
from sklearn import metrics
from sklearn.datasets.samples_generator import make_blobs
from sklearn.preprocessing import StandardScaler
from scipy import stats
from scipy import cluster
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
alpha_std = 0.0178
fe_std = 0.007
num_centers = 1000
n_samples = 1000
```

```python
x, y = make_blobs(n_samples=n_samples, centers=num_centers, n_features=2,
                  cluster_std=[alpha_std, fe_std], center_box=[-2, 1])
```

```python
for center in xrange(num_centers):
    plt.plot(x[y == center, 0], x[y == center, 1], '.')
```
